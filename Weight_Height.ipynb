```python
inputtens = torch.tensor(arr)
inputtens
```

```python
outputtens = torch.tensor(arr1)
outputtens
```

```python
w = torch.randn(1,1,requires_grad = True)
b = torch.randn(1000,1, requires_grad=True)
print(w);print(b)
```

```python
def model(x):
    return x@w.t()+b
```

```python
pred = model(inputtens)
pred
```

```python
def mse(t1,t2):
    diff = t1 - t2
    return torch.sum(diff*diff) / torch.numel(t1)
```

```python
loss = mse(pred, outputtens)
loss
```

```python
import torch.nn as nn
```

```python
from torch.utils.data import TensorDataset
```

```python
#Define dataset
train_ds = TensorDataset(inputtens, outputtens)
train_ds[0:3]
```

```python
from torch.utils.data import DataLoader
```

```python
batch_size = 25
train_dl = DataLoader(train_ds,batch_size,shuffle =True)
```

```python
for xb,yb in train_dl:
    print(xb)
    print(yb)
    break
```

```python
#define model 
model = nn.Linear(1,1)
print(model.weight)
print(model.bias)
```

```python
list(model.parameters())
```

```python
pred = model(inputtens)
pred
```

```python
import torch.nn.functional as F
```

```python
loss_fn = F.mse_loss
```

```python
loss = loss_fn(model(inputtens),outputtens)
loss
```

```python
opt = torch.optim.SGD(model.parameters(),lr=1e-5)
```

```python
def fit(num_epochs, model, loss_fn, opt, train_dl):
    for epoch in range(num_epochs):
        for xb,yb in train_dl:
            pred = model(xb)
            loss = loss_fn(pred,yb)
            loss.backward()
            opt.step()
            opt.zero_grad()
        if(epoch+1)%10 == 0:
            print('Epoch [{}/{} Loss: {:.4f}'.format(epoch+1, num_epochs,loss.item()))
```

```python
fit(100,model,loss_fn,opt,train_dl)
```
